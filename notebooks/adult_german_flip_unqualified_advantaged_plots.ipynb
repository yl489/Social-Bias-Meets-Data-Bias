```python
import numpy as np
import matplotlib.pyplot as plt
import sys
plt.rcParams.update({'font.size': 15, 'lines.linewidth':3.5})
```

## Load Adult

```python
A_DP = np.load('./adult_german_results/adult_DP_0.01_dp_2_10.npz')
A_TPR = np.load('./adult_german_results/adult_TPR_0.01_tpr_2_10.npz')
A_FPR = np.load('./adult_german_results/adult_FPR_0.01_fpr_2_10.npz')
A_EO = np.load('./adult_german_results/adult_EO_0.01_eo_2_10.npz')
# unconstrained
A_acc = np.concatenate((A_DP['acc_overall'], A_TPR['acc_overall'], A_FPR['acc_overall'], A_EO['acc_overall']), axis=1)
A_acc_mu = np.mean(A_acc, axis=1)
A_acc_std = np.std(A_acc, axis=1)
# A_DP
A_DP_acc_mitigated = A_DP['acc_overall_mitigated']
A_DP_acc_mitigated_mu = np.mean(A_DP_acc_mitigated, axis=1)
A_DP_acc_mitigated_std = np.std(A_DP_acc_mitigated, axis=1)
A_DP_diff_mitigated = A_DP['diff_mitigated']
A_DP_diff_mitigated_mu = np.mean(A_DP_diff_mitigated, axis=1)
A_DP_diff_mitigated_std = np.std(A_DP_diff_mitigated, axis=1)
# A_TPR
A_TPR_acc_mitigated = A_TPR['acc_overall_mitigated']
A_TPR_acc_mitigated_mu = np.mean(A_TPR_acc_mitigated, axis=1)
A_TPR_acc_mitigated_std = np.std(A_TPR_acc_mitigated, axis=1)
A_TPR_diff_mitigated = A_TPR['diff_mitigated']
A_TPR_diff_mitigated_mu = np.mean(A_TPR_diff_mitigated, axis=1)
A_TPR_diff_mitigated_std = np.std(A_TPR_diff_mitigated, axis=1)
# A_FPR
A_FPR_acc_mitigated = A_FPR['acc_overall_mitigated']
A_FPR_acc_mitigated_mu = np.mean(A_FPR_acc_mitigated, axis=1)
A_FPR_acc_mitigated_std = np.std(A_FPR_acc_mitigated, axis=1)
A_FPR_diff_mitigated = A_FPR['diff_mitigated']
A_FPR_diff_mitigated_mu = np.mean(A_FPR_diff_mitigated, axis=1)
A_FPR_diff_mitigated_std = np.std(A_FPR_diff_mitigated, axis=1)
# A_EO
A_EO_acc_mitigated = A_EO['acc_overall_mitigated']
A_EO_acc_mitigated_mu = np.mean(A_EO_acc_mitigated, axis=1)
A_EO_acc_mitigated_std = np.std(A_EO_acc_mitigated, axis=1)
A_EO_diff_mitigated = A_EO['diff_mitigated']
A_EO_diff_mitigated_mu = np.mean(A_EO_diff_mitigated, axis=1)
A_EO_diff_mitigated_std = np.std(A_EO_diff_mitigated, axis=1)
```

## Load German

```python
G_DP = np.load('./adult_german_results/german_DP_0.01_dp_2_30.npz')
G_TPR = np.load('./adult_german_results/german_TPR_0.01_tpr_2_30.npz')
G_FPR = np.load('./adult_german_results/german_FPR_0.01_fpr_2_30.npz')
G_EO = np.load('./adult_german_results/german_EO_0.01_eo_2_30.npz')
# unconstrained
G_acc = np.concatenate((G_DP['acc_overall'], G_TPR['acc_overall'], G_FPR['acc_overall'], G_EO['acc_overall']), axis=1)
G_acc_mu = np.mean(G_acc, axis=1)
G_acc_std = np.std(G_acc, axis=1)
# G_DP
G_DP_acc_mitigated = G_DP['acc_overall_mitigated']
G_DP_acc_mitigated_mu = np.mean(G_DP_acc_mitigated, axis=1)
G_DP_acc_mitigated_std = np.std(G_DP_acc_mitigated, axis=1)
G_DP_diff_mitigated = G_DP['diff_mitigated']
G_DP_diff_mitigated_mu = np.mean(G_DP_diff_mitigated, axis=1)
G_DP_diff_mitigated_std = np.std(G_DP_diff_mitigated, axis=1)
# G_TPR
G_TPR_acc_mitigated = G_TPR['acc_overall_mitigated']
G_TPR_acc_mitigated_mu = np.mean(G_TPR_acc_mitigated, axis=1)
G_TPR_acc_mitigated_std = np.std(G_TPR_acc_mitigated, axis=1)
G_TPR_diff_mitigated = G_TPR['diff_mitigated']
G_TPR_diff_mitigated_mu = np.mean(G_TPR_diff_mitigated, axis=1)
G_TPR_diff_mitigated_std = np.std(G_TPR_diff_mitigated, axis=1)
# G_FPR
G_FPR_acc_mitigated = G_FPR['acc_overall_mitigated']
G_FPR_acc_mitigated_mu = np.mean(G_FPR_acc_mitigated, axis=1)
G_FPR_acc_mitigated_std = np.std(G_FPR_acc_mitigated, axis=1)
G_FPR_diff_mitigated = G_FPR['diff_mitigated']
G_FPR_diff_mitigated_mu = np.mean(G_FPR_diff_mitigated, axis=1)
G_FPR_diff_mitigated_std = np.std(G_FPR_diff_mitigated, axis=1)
# G_EO
G_EO_acc_mitigated = G_EO['acc_overall_mitigated']
G_EO_acc_mitigated_mu = np.mean(G_EO_acc_mitigated, axis=1)
G_EO_acc_mitigated_std = np.std(G_EO_acc_mitigated, axis=1)
G_EO_diff_mitigated = G_EO['diff_mitigated']
G_EO_diff_mitigated_mu = np.mean(G_EO_diff_mitigated, axis=1)
G_EO_diff_mitigated_std = np.std(G_EO_diff_mitigated, axis=1)
```

## Visualize

```python
x = np.linspace(0, 0.5, 6)
fig, ax = plt.subplots(2,2,figsize=(10,8))
fs = 20
ms = 10
alpha = 0.15
# adult acc
ax[0,0].plot(x, A_acc_mu, color='navy', marker='^', markersize=ms)
ax[0,0].fill_between(x, y1=(A_acc_mu-A_acc_std), y2=(A_acc_mu+A_acc_std), color='navy', alpha=alpha)
ax[0,0].plot(x, A_DP_acc_mitigated_mu, color='orange', marker='o', markersize=ms)
ax[0,0].fill_between(x, y1=(A_DP_acc_mitigated_mu-A_DP_acc_mitigated_std), 
    y2=(A_DP_acc_mitigated_mu+A_DP_acc_mitigated_std), color='orange', alpha=alpha)
ax[0,0].plot(x, A_TPR_acc_mitigated_mu, color='cyan', marker='v', markersize=ms)
ax[0,0].fill_between(x, y1=(A_TPR_acc_mitigated_mu-A_TPR_acc_mitigated_std), 
    y2=(A_TPR_acc_mitigated_mu+A_TPR_acc_mitigated_std), color='cyan', alpha=alpha)
ax[0,0].plot(x, A_FPR_acc_mitigated_mu, color='purple', marker='s', markersize=ms)
ax[0,0].fill_between(x, y1=(A_FPR_acc_mitigated_mu-A_FPR_acc_mitigated_std), 
    y2=(A_FPR_acc_mitigated_mu+A_FPR_acc_mitigated_std), color='purple', alpha=alpha)
ax[0,0].plot(x, A_EO_acc_mitigated_mu, color='green', marker='D', markersize=ms)
ax[0,0].fill_between(x, y1=(A_EO_acc_mitigated_mu-A_EO_acc_mitigated_std), 
    y2=(A_EO_acc_mitigated_mu+A_EO_acc_mitigated_std), color='green', alpha=alpha)
# adult fairness
ax[1,0].plot(x, A_DP_diff_mitigated_mu, color='orange', marker='o', markersize=ms)
ax[1,0].fill_between(x, y1=(A_DP_diff_mitigated_mu-A_DP_diff_mitigated_std), 
    y2=(A_DP_diff_mitigated_mu+A_DP_diff_mitigated_std), color='orange', alpha=alpha)
ax[1,0].plot(x, A_TPR_diff_mitigated_mu, color='cyan', marker='v', markersize=ms)
ax[1,0].fill_between(x, y1=(A_TPR_diff_mitigated_mu-A_TPR_diff_mitigated_std), 
    y2=(A_TPR_diff_mitigated_mu+A_TPR_diff_mitigated_std), color='cyan', alpha=alpha)
ax[1,0].plot(x, A_FPR_diff_mitigated_mu, color='purple', marker='s', markersize=ms)
ax[1,0].fill_between(x, y1=(A_FPR_diff_mitigated_mu-A_FPR_diff_mitigated_std), 
    y2=(A_FPR_diff_mitigated_mu+A_FPR_diff_mitigated_std), color='purple', alpha=alpha)
ax[1,0].plot(x, A_EO_diff_mitigated_mu, color='green', marker='D', markersize=ms)
ax[1,0].fill_between(x, y1=(A_EO_diff_mitigated_mu-A_EO_diff_mitigated_std), 
    y2=(A_EO_diff_mitigated_mu+A_EO_diff_mitigated_std), color='green', alpha=alpha)
# german acc
ax[0,1].plot(x, G_acc_mu, label='Unconstrained', color='navy', marker='^', markersize=ms)
ax[0,1].fill_between(x, y1=(G_acc_mu-G_acc_std), y2=(G_acc_mu+G_acc_std), color='navy', alpha=alpha)
ax[0,1].plot(x, G_DP_acc_mitigated_mu, color='orange', marker='o', markersize=ms)
ax[0,1].fill_between(x, y1=(G_DP_acc_mitigated_mu-G_DP_acc_mitigated_std), 
    y2=(G_DP_acc_mitigated_mu+G_DP_acc_mitigated_std), color='orange', alpha=alpha)
ax[0,1].plot(x, G_TPR_acc_mitigated_mu, color='cyan', marker='v', markersize=ms)
ax[0,1].fill_between(x, y1=(G_TPR_acc_mitigated_mu-G_TPR_acc_mitigated_std), 
    y2=(G_TPR_acc_mitigated_mu+G_TPR_acc_mitigated_std), color='cyan', alpha=alpha)
ax[0,1].plot(x, G_FPR_acc_mitigated_mu, color='purple', marker='s', markersize=ms)
ax[0,1].fill_between(x, y1=(G_FPR_acc_mitigated_mu-G_FPR_acc_mitigated_std), 
    y2=(G_FPR_acc_mitigated_mu+G_FPR_acc_mitigated_std), color='purple', alpha=alpha)
ax[0,1].plot(x, G_EO_acc_mitigated_mu, color='green', marker='D', markersize=ms)
ax[0,1].fill_between(x, y1=(G_EO_acc_mitigated_mu-G_EO_acc_mitigated_std), 
    y2=(G_EO_acc_mitigated_mu+G_EO_acc_mitigated_std), color='green', alpha=alpha)
# german fairness
ax[1,1].plot(x, G_DP_diff_mitigated_mu, label='DP', color='orange', marker='o', markersize=ms)
ax[1,1].fill_between(x, y1=(G_DP_diff_mitigated_mu-G_DP_diff_mitigated_std), 
    y2=(G_DP_diff_mitigated_mu+G_DP_diff_mitigated_std), color='orange', alpha=alpha)
ax[1,1].plot(x, G_TPR_diff_mitigated_mu, label='TPR', color='cyan', marker='v', markersize=ms)
ax[1,1].fill_between(x, y1=(G_TPR_diff_mitigated_mu-G_TPR_diff_mitigated_std), 
    y2=(G_TPR_diff_mitigated_mu+G_TPR_diff_mitigated_std), color='cyan', alpha=alpha)
ax[1,1].plot(x, G_FPR_diff_mitigated_mu, label='FPR', color='purple', marker='s', markersize=ms)
ax[1,1].fill_between(x, y1=(G_FPR_diff_mitigated_mu-G_FPR_diff_mitigated_std), 
    y2=(G_FPR_diff_mitigated_mu+G_FPR_diff_mitigated_std), color='purple', alpha=alpha)
ax[1,1].plot(x, G_EO_diff_mitigated_mu, label='EO', color='green', marker='D', markersize=ms)
ax[1,1].fill_between(x, y1=(G_EO_diff_mitigated_mu-G_EO_diff_mitigated_std), 
    y2=(G_EO_diff_mitigated_mu+G_EO_diff_mitigated_std), color='green', alpha=alpha)

ax[0,0].spines['right'].set_visible(False)
ax[0,0].spines['top'].set_visible(False)
ax[1,0].spines['right'].set_visible(False)
ax[1,0].spines['top'].set_visible(False)
ax[0,1].spines['right'].set_visible(False)
ax[0,1].spines['top'].set_visible(False)
ax[1,1].spines['right'].set_visible(False)
ax[1,1].spines['top'].set_visible(False)
ax[0,0].set_xlim(0,0.5)
ax[0,0].grid()
ax[1,0].set_xlim(0,0.5)
ax[1,0].grid()
ax[0,1].set_xlim(0,0.5)
ax[0,1].grid()
ax[1,1].set_xlim(0,0.5)
ax[1,1].grid()
ax[0,0].set_title('Adult', fontsize=fs)
ax[0,1].set_title('German', fontsize=fs)
ax[0,0].set_ylabel('Accuracy', fontsize=fs)
ax[1,0].set_ylabel('Fairness Violation', fontsize=fs)
ax[1,0].set_xlabel('P(flipping unqualified male)', fontsize=fs)
ax[1,1].set_xlabel('P(flipping unqualified above 30)', fontsize=fs)
fig.legend(bbox_to_anchor=(0.05,-0.05,0.9,0.2), loc="lower left", 
    mode="expand", borderaxespad=0, ncol=5,frameon=False, fontsize=fs)
plt.tight_layout()
```
